```python
# using SMOTE for oversampling
from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(x, y)

# Print the results
print("Original data:")
# print(x.info(0))
print(y.value_counts())

print("\_Resampled data:")
# print(x_smote.info())
print(y_smote.value_counts())

from sklearn.model_selection import train_test_split
# Splitting data menjadi training dan test dataset sesuai dengan rasio-nya:
# Training : 80%, Test : 20%
# Menggunakan fungsi train_test_split()
# dan memasukkan hasilnya ke variabel masing-masing
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.2, random_state=42)
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
print("Oversampling (SMOTE)")

# Yak ini SVM yak
model_svm = svm.SVC(kernel='linear')
svm_save = model_svm
model_svm.fit(x_train, y_train)
svm_pred = model_svm.predict(x_test)

# Create confusion matrix
cm = confusion_matrix(y_test, svm_pred)

# Extract values from confusion matrix
tn = cm[0, 0]  # True negatives
fp = cm[0, 1]  # False positives

svm_precision = precision_score(y_test, svm_pred)
svm_specificity = tn / (tn + fp)
svm_sensitivity = recall_score(y_test, svm_pred)
svm_f1_score = f1_score(y_test, svm_pred, average='weighted')

print("--- SVM ---")
print("Precision:", svm_precision*100, "%")
print("Sensitivity:", svm_sensitivity*100, "%")
print("Specificity:", svm_specificity*100, "%")
print("F1 Score:", svm_f1_score*100, "%")

print(confusion_matrix(y_test, svm_pred))
print(classification_report(y_test, svm_pred))




# Algo, algo apa yang deket kita trus hidupnya acak-acakan?
model_rf = RandomForestClassifier()
rf_save = model_rf
model_rf.fit(x_train, y_train)
rf_pred = model_rf.predict(x_test)

# Create confusion matrix
cm = confusion_matrix(y_test, rf_pred)

# Extract values from confusion matrix
tn = cm[0, 0]  # True negatives
fp = cm[0, 1]  # False positives

rf_precision = precision_score(y_test, rf_pred)
rf_specificity = tn / (tn + fp)
rf_sensitivity = recall_score(y_test, rf_pred)
rf_f1_score = f1_score(y_test, rf_pred, average='weighted')

print("--- RF ---")
print("Precision:", rf_precision*100, "%")
print("Sensitivity:", rf_sensitivity*100, "%")
print("Specificity:", rf_specificity*100, "%")
print("F1 Score:", rf_f1_score*100, "%")

print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))



# XGBoost
model_xgb = xgb.XGBClassifier()
xgb_save = model_xgb
model_xgb.fit(x_train, y_train)
xgb_pred = model_xgb.predict(x_test)

# Create confusion matrix
cm = confusion_matrix(y_test, xgb_pred)

# Extract values from confusion matrix
tn = cm[0, 0]  # True negatives
fp = cm[0, 1]  # False positives

xgb_precision = precision_score(y_test, xgb_pred)
xgb_specificity = tn / (tn + fp)
xgb_sensitivity = recall_score(y_test, xgb_pred)
xgb_f1_score = f1_score(y_test, xgb_pred, average='weighted')

print("--- XGBoost ---")
print("Precision:", xgb_precision*100, "%")
print("Sensitivity:", xgb_sensitivity*100, "%")
print("Specificity:", xgb_specificity*100, "%")
print("F1 Score:", xgb_f1_score*100, "%")

print(confusion_matrix(y_test, xgb_pred))
print(classification_report(y_test, xgb_pred))



# Algo yang suka nyari tetangga baru
model_knn = KNeighborsClassifier()
knn_save = model_knn
model_knn.fit(x_train, y_train)
knn_pred = model_knn.predict(x_test)

# Create confusion matrix
cm = confusion_matrix(y_test, knn_pred)

# Extract values from confusion matrix
tn = cm[0, 0]  # True negatives
fp = cm[0, 1]  # False positives

knn_precision = precision_score(y_test, knn_pred)
knn_specificity = tn / (tn + fp)
knn_sensitivity = recall_score(y_test, knn_pred)
knn_f1_score = f1_score(y_test, knn_pred, average='weighted')

print("--- KNN ---")
print("Precision:", knn_precision*100, "%")
print("Sensitivity:", knn_sensitivity*100, "%")
print("Specificity:", knn_specificity*100, "%")
print("F1 Score:", knn_f1_score*100, "%")

print(confusion_matrix(y_test, knn_pred))
print(classification_report(y_test, knn_pred))
```
